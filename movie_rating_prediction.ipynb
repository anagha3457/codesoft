```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

handling missing values

```python
df['Duration'] = df['Duration'].str.replace('min', '', regex=False)
df['Duration'] = pd.to_numeric(df['Duration'], errors='coerce')
```

```python
df.head()
```

```python
df.info()
```

```python
df = df.dropna(subset=['Rating'])
```

```python
df['Votes'] = df['Votes'].str.replace(',', '', regex=False)
df['Votes'] = pd.to_numeric(df['Votes'], errors='coerce')
```

```python
df['Duration'] = df['Duration'].fillna(df['Duration'].mean())
```

```python
for col in ['Director', 'Actor 1', 'Actor 2', 'Actor 3']:
    df[col] = df[col].fillna(df[col].mode()[0])
```

```python
df.info()
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.shape
```

string ----> numeric

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
df["Director"]= le.fit_transform(df['Director'])
```

```python
df["Actor 1"]= le.fit_transform(df['Actor 1'])
```

```python
df["Actor 2"]= le.fit_transform(df['Actor 2'])
```

```python
df["Actor 3"]= le.fit_transform(df['Actor 3'])
```

```python
df['Genre'] = df['Genre'].fillna('Unknown')
df['Genre_list'] = df['Genre'].str.split(', ')
```

```python
from sklearn.preprocessing import MultiLabelBinarizer

mlb = MultiLabelBinarizer()
genre_encoded = pd.DataFrame(mlb.fit_transform(df['Genre_list']), columns=mlb.classes_)

# Reset index to align with original df
genre_encoded.index = df.index
```

```python
df = df.drop(columns=['Genre', 'Genre_list'])
df = pd.concat([df, genre_encoded], axis=1)
```

```python
df.head()
```

```python
df.info()
```

model training and eval

```python
X = df.drop( columns = ['Rating' , 'Name', 'Year'])
Y = df['Rating']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split( X, Y, test_size = 0.2, random_state = 42)
```

```python
model = LinearRegression()
```

```python
model.fit(X_train, Y_train)
```

```python
X_train_pred = model.predict(X_train)
```

```python
train_accuracy = mean_squared_error(Y_train, X_train_pred)
```

```python
print("accuracy on training data = ", train_accuracy )
```

```python
X_test_pred = model.predict(X_test)
```

```python
test_accuracy = mean_squared_error(Y_test, X_test_pred)
```

```python
print("accuracy on test data = ", test_accuracy)
```
